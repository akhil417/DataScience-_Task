```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.scatter(x,y)
plt.show()
```

```python
m=0
c=0
Yp=m*x+c
```

```python
#LR = Learning Rate
LR=0.0001
iteration = 1000
n = float(len(x))
print(n)
 
for i in range(iteration):
    Yp = (m*x)+c
    D_m = (-2/n)*sum(x * (y - Yp))
    D_c = (-2/n)*sum(y - Yp)
    m = m - (LR*D_m) 
    c = c - (LR*D_c)
    print("Slope: ",m)
    print("Intercept: ",c)
    
```

```python
Yp=m*x+c
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(Yp),max(Yp)],color = 'red')
plt.show()
```
